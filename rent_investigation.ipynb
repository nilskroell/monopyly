```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rents = [[2, 10, 30, 90, 160, 250],
 [4, 20, 60, 180, 320, 450],
 [6, 30, 90, 270, 400, 550],
 []]
```

```python
rents = np.array(rents)
```

```python
rents_norm = rents / np.tile(np.reshape(rents[:,0].T, (rents.shape[0], 1)), rents.shape[1])
```

```python
plt.plot(rents_norm.T)
```
